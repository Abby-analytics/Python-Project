```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df_us = df[df['job_country']=="United States"].copy()
df_us['job_posted_month'] = df_us['job_posted_date'].dt.strftime('%B')
df_us_pivot = df_us.pivot_table(index='job_posted_month', columns='job_title_short', aggfunc='size')
df_us_pivot = df_us_pivot.reset_index()
df_us_pivot['month_no'] = pd.to_datetime(df_us_pivot['job_posted_month'], format='%B').dt.month
df_us_pivot = df_us_pivot.sort_values('month_no')
df_us_pivot = df_us_pivot.set_index('job_posted_month')



df_us_pivot
```

```python
df_us_pivot = df_us_pivot.drop(columns='month_no')
```

```python
if 'month_no' in df_us_pivot.columns:
    df_us_pivot = df_us_pivot.drop(columns='month_no')
```

```python
df_us_pivot
```

```python
df_new_job = pd.read_csv("https://lukeb.co/software_csv")
```

```python
df_new_job.set_index('job_posted_month')
```

```python
df_us_merge = df_us_pivot.merge(df_new_job, on='job_posted_month')
```

```python
top_5 = (df_us_merge
.select_dtypes(include='number')
.sum()
.sort_values(ascending=False)
.head().index.to_list())
top_5
```

```python
df_us_merge[top_5].plot(kind='line')
plt.title('Monthly job Postings for Top Tech jobs in the us')
plt.xlabel('2023')
plt.ylabel('Job Count')
plt.ylim(0, 20000)
plt.legend()
plt.show()
```
